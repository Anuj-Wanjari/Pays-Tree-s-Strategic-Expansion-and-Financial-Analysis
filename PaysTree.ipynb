# Revenue Growth

```python
import matplotlib.pyplot as plt

years = ["FY22", "FY23", "FY24"]
revenue = [123, 179.5, 236]

plt.figure(figsize=(8, 4))
plt.plot(years, revenue, marker='o', linestyle='-', color='b')
plt.title("Pays Tree Revenue Growth")
plt.xlabel("Financial Year")
plt.ylabel("Revenue (in crore rupees)")
plt.grid(True)
plt.show()
```

# Profit Trends

```python
years = ["FY21", "FY22", "FY23"]
profits = [6.6, 23, 0.95]

plt.figure(figsize=(8, 4))
plt.bar(years, profits, color=['red', 'green', 'blue'])
plt.title("Pays Tree Profit Trends")
plt.xlabel("Financial Year")
plt.ylabel("Profit (in crore rupees)")
plt.show()
```

# Transaction Volume

```python
years = ["FY22", "Current Year"]
gtv = [28000, 70000]  # in crore rupees

plt.figure(figsize=(8, 4))
plt.plot(years, gtv, marker='o', linestyle='-', color='g')
plt.title("Pays Tree Gross Transaction Value (GTV)")
plt.xlabel("Financial Year")
plt.ylabel("GTV (in crore rupees)")
plt.grid(True)
plt.show()
```

### Step-by-Step Calculation for FY23

> Add blockquote



1. **Identify the known values:**
   - Revenue for FY22 (\( R_{FY22} \)) = ₹123 crores
   - Revenue for FY24 (\( R_{FY24} \)) = ₹236 crores

2. **Calculate the difference between FY24 and FY22 revenues:**
   \[ R_{FY24} - R_{FY22} = 236 - 123 = 113 \]

3. **Since FY23 is exactly halfway between FY22 and FY24, we divide the difference by 2:**
   \[ \frac{113}{2} = 56.5 \]

4. **Add this halfway point to the FY22 revenue to get the estimated FY23 revenue:**
   \[ R_{FY23} = 123 + 56.5 = 179.5 \]

Therefore, the estimated revenue for FY23 is ₹179.5 crores.
